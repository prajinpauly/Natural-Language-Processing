### **Question Answering System**

```python
from difflib import get_close_matches
```

```python
# Function to find the similarity between question asked by used and predefined questions
def find_similar_question(question, questions):
  similar_questions = get_close_matches(question, questions, n=1, cutoff=0.8)
  if similar_questions:
    return similar_questions[0]
  else:
    return None
```

```python
import re
```

```python
# Function to extract reviews for the questions asked (matching keywords)
def search_reviews(question, keywords, reviews):
  # Compile keywords into a regular expression pattern
  pattern = re.compile(r'\b(?:' + '|'.join(keywords) + r')\b', flags=re.IGNORECASE)

  # Search reviews for sentences containing the keywords
  matching_sentences = []
  for review in reviews:
    sentences = review.split('.')
    for sentence in sentences:
      if re.search(pattern, sentence):
        matching_sentences.append(sentence.strip())

  return matching_sentences
```

```python
# Function to extract keywords and put all functions together and handle possible errors
def answer(question, questions_keywords, reviews):
  similar_question = find_similar_question(question, questions_keywords.keys())
  if similar_question:
    keywords = questions_keywords[similar_question]
    matching_sentences = search_reviews(similar_question, keywords, reviews)
    if matching_sentences:
      return keywords, matching_sentences
    else:
      return 'Sorry, no relevant information found in the reviews.'
  else:
    return 'Sorry, could not find a similar question in the dataset.'
```

```python
from transformers import BertTokenizer, BertModel
import torch
import numpy as np
from scipy.spatial.distance import cosine

# Initialize tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Function to convert text into BERT embeddings
def get_bert_embedding(text):
    # Encode the text using tokenizer
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    # Get outputs from the model
    outputs = model(**inputs)
    # Get the embeddings for the [CLS] token (first token), which represents the sentence
    embeddings = outputs.last_hidden_state[:, 0, :]
    return embeddings.detach()

# Function to calculate cosine similarity
def calculate_similarity(embedding1, embedding2):
    # Ensure the embeddings are 1-D by squeezing the batch dimension
    emb1 = embedding1.squeeze(0).numpy()
    emb2 = embedding2.squeeze(0).numpy()
    # Compute cosine similarity
    return 1 - cosine(emb1, emb2)

# Example question and answers
question = question1
answers = answer1[:10]
similarity_score = []

# Get embeddings for the question
question_embedding = get_bert_embedding(question)

# Calculate similarities and display them
print("Semantic Similarities:")
for answer in answers:
    answer_embedding = get_bert_embedding(answer)
    similarity = calculate_similarity(question_embedding, answer_embedding)
    print(f"Question: {question}")
    print(f"Answer: {answer}")
    print(f"Similarity Score: {similarity:.3f}")
    print("-" * 50)
    similarity_score.append(similarity)

final = np.array(similarity_score).mean()

print(final)
```
